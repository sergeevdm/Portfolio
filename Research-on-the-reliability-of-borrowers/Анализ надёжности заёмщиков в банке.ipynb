### Шаг 3. Ответьте на вопросы

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
# запустим наш любимый цикл для уточнения значения задолжности debt по категориям
def debt_means(debt):
    if debt == 1:
        return 'есть задолжности'
    if debt == 0:
        return 'нет задолжностей'
data_drop_dupl['debt_means'] = data_drop_dupl['debt'].apply(debt_means)
# проверим абсолютную способность к выплатам в зависимости от наличия детей 
data_drop_dupl.groupby('debt_means')['children_category'].value_counts()                                
```

```python
# найдем процент должников "с детьми" и "без" в % 
children_debt = data_drop_dupl.groupby(['children_category']).agg({'debt' : ['sum','count']})
children_debt['part'] = (children_debt['debt']['sum'] / children_debt['debt']['count']) * 100
children_debt.sort_values(by = 'part', ascending = False)
```

### Вывод

По % задожность слабо прослеживается. При этом, тех у кого "нет детей" берут кредиты или сразу отдают в большинстве. С детьми люди более состоятельные и в кредитах не нуждаются
